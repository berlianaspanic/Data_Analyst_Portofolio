###3.Data Manipulation

####a. Change Data Structure

```python
#show data structure
df
```

Because the structure of dataframe is not efficient to use for analysis. We can repair that structur of dataframe.

```python
#repair dataframe structure
df_cc = df[df['Target'].isin(['ConfirmedCases'])] #get data with filtering for `Target` variabel is 'ConfirmedCases'
df_cc = df_cc.drop(columns=['Target']) #drop `Target` variabel to repair be column
df_cc = df_cc.reset_index(drop=True) #make new index for dataframe has filtered
df_cc.rename(columns={"TargetValue":"Confirmed Cases"}, inplace=True) #rename `TargetValue` to be show count of `Confirmed Cases`
print(df_cc.shape) #show dataframe shape
df_cc.head() #show dataframe head
```

```python
#repair dataframe structure
df_f = df[df['Target'].isin(['Fatalities'])] #get data with filtering for `Target` variabel is 'FatalitiesCases'
df_f = df_f.drop(columns=['Target']) #drop `Target` variabel to repair be column
df_f = df_f.reset_index(drop=True) #make new index for dataframe has filtered
df_f.rename(columns={"TargetValue":"Fatalities Cases"}, inplace=True) #rename `TargetValue` to be show count of `Fatalities Cases`
print(df_f.shape) #show dataframe shape
df_f.head() #show dataframe head
```

```python
#repair dataframe structure
cc_value = df_cc['Confirmed Cases'] #get 'Confirmed Cases' in df_cc dataframe
df_f['Confirmed Cases'] = cc_value #make new column in df_f dataframe filled the 'Confirmed Cases' value
print(df_f.shape) #show df_f dataframe has repaired
df_f #show df_f dataframe
```

The new structure of dataframe make the dataframe has a new count of rows. dataframe has 484820 rows now.

```python
#download result data manipulation
df_f.to_csv(index=False)
compression_opts = dict(method='zip',
                        archive_name='global_covid.xlsx')  
df_f.to_csv('out.zip', index=False,
          compression=compression_opts) 
from pathlib import Path  
filepath = Path('/content/drive/MyDrive/Colab Notebooks/global_covid.xlsx')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
df_f.to_csv(filepath)  
import os  
os.makedirs('/content/drive/MyDrive/Colab Notebooks/', exist_ok=True)  
df_f.to_csv('/content/drive/MyDrive/Colab Notebooks/global_covid.xlsx')  
```

####b. Aggregation Manipulation For Looker

Because looker studio can not use sum_distict and `Population` variabel has a same value in record. So, to show the count of `Population` we have use different way. One way, we can use average function. So, we have known that frequently of duplicated value.

```python
#show value count to know minimum looping for population and country_ragion variabel
print(df_f['Population'].value_counts())
print(df_f['Country_Region'].value_counts())
```

Based on above output, we known that each unique has 140 duplicated record which is value_counts of `Date` variabel. So, we use 140 in counting real `Population`.

```python
#show sum population variabel for each country_region
agg2 = df_f.groupby('Country_Region')['Population'].sum()/140
agg2
agg3 = pd.DataFrame(agg2)
agg3['Population'].sum()
```

Based on above output, we known that counting of `Population` used the `Average` way is usefull because result rigth value for show sum of all `Population`.
