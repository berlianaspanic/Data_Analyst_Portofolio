### Create Dataframe

Membuat 2 dataframe dari kumpulan list data (dictionary).
code: ***namadataframe=pd.Dataframe(dictionary)***

Dataframe pertama bernama 'df1' yang terdiri dari 2 Variabel/kolom yaitu variabel 'key' dan variabel 'A' di mana setiap variabel memiliki 6 nilai data.

Dataframe kedua bernama 'df2' yang terdiri dari 2 variabel/kolom yaitu variabel 'key' dan variabel 'B' di mana setiap variabel memiliki 3 nilai data.

```python
# Dataframe 1
df1 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
df1
```

```python
# Dataframe 2
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})
df2
```

### Create a dataframe (df3) which consists both dataframe inner-joined by column "Key" (5 pts)

Membuat dataframe bernama 'df3' yang merupakan hasil inner join dari df1 dengan df2. Terdapat beberapa code yang dapat digunakan:
1. ***namadataframehasiljoin = pd.merge(namadataframe1,namadataframe2,how="inner", on='namakolomvariabelirisan')***
2. ***namadataframehasiljoin = namadataframe1.merge(namadataframe2,how="inner", on='namakolomvariabelirisan')***
3. code ketiga ini digunakan apabila nama kolom yang digunakan sebagai irisan berbeda
***namadataframehasiljoin = pd.merge(namadataframe1,namadataframe2,how="inner", left_on='namakolomvariabelirisan1', rigth_on='namakolomvariabelirisan12')***

Pada soal ini, kolom variabel 'key' merupakan kolom variabel irisannya. Inner Join akan menghasilkan dataframe hasil irisan dari 2 dataframe sebelumnya. Hasil df3 akan membentuk dataframe yang terdiri dari 3 variabel kolom yaitu kolom irisan 'key', kolom 'A' yang berasalah dari df1, dan kolom 'B' yang berasal dari df2. Serta terdiri dari 3 baris nilai data karena irisan df1 dengan df2 pada kolom 'key' hanya memiliki 3 baris nilai data.

```python
#code goes here
#Inner-join df1 with df2
df3 = pd.merge(df1,df2,how="inner", on='key')
df3
```

### Create a dataframe (df4) which consists of the result of df1 and df2 left-joined by column "Key" (10pts)

Membuat dataframe bernama 'df3' yang merupakan hasil left join dari df1 dengan df2. Terdapat beberapa code yang dapat digunakan.
1. ***namadataframehasiljoin = pd.merge(namadataframe1,namadataframe2,how="left", on='namakolomvariabelgabungan')***
2. ***namadataframehasiljoin = namadataframe1.merge(namadataframe2,how="left", on='namakolomvariabelgabungan')***
3. code ketiga ini digunakan apabila nama kolom yang akan digabungkan berbeda
***namadataframehasiljoin = pd.merge(namadataframe1,namadataframe2,how="left", left_on='namakolomvariabelirisan1', rigth_on='namakolomvariabelirisan12')***

Pada soal ini, kolom variabel 'key' adalah kolom yang akan digabungkan. Left Join akan menghasilkan dataframe hasil penggabungan dari 2 dataframe sebelumnya berdasarkan seluruh nilai data pada dataframe utama dalam hal ini 'df1'. Left join biasanya dapat memunculkan adanya data 'NaN' apabila nilai data dari kolom variabel gabungan antara 2 dataframe berbeda. 
Sehingga df4 akan menghasilkan dataframe yang berisi seluruh nilai data df1 dan irisan data df2 berdasarkan kolom variael gabungan 'key' pada df1. Di mana terdiri dari 3 kolom variabel yaitu kolom gabungan 'key', kolom 'A' dari df1, dan kolom 'B' dari df2, Serta terdiri dari 6 baris nilai data. Kolom variabel gabungan 'key' akan memiliki 6 nilai data sesuai jumlah nilai data kolom key df1. Kolom 'A' akan memiliki nilai data yang terisi penuh karena berasal dari df1. Sementara pada kolom 'B' akan terdapat 3 nilai data NaN karena kolom B berasal dari df2 yang hanya memiliki 3 baris nilai data.

```python
#code goes here
#Left-join df1 with df2
df4 = pd.merge(df1,df2,how="left", on="key")
df4
```

Create a dictionary and convert it to a dataframe called "Customer_df" (15 pts)

Membuat dataframe dari kumpulan list data dapat dilakukan dengan menjadikan kumpulan list data tersebut menjadi dictionary dengan menggunakan code: ***namadictionary= {list1, list2, dst}*** dan menggunakan code: ***namadataframe=pd.Dataframe.from_dict(namadictionary)*** untuk menjadikannya sebuah dataframe.

Dataframe "Customer_dict" akan terdiri dari 3 kolom variabel yaitu name, age, dan gender serta terdiri dari 3 baris nilai data karena setiap list yang ada berisi 3 nilai data.



```python
#code goes here
#Create Dictionary Datatype
Customer_dict={
    'name':['Anna','Jason','Cindy'],
    'age':['22','23','21'],
    'gender':['F','M','F']
}
#Convert Dictionary to Dataframe
Customer_df = pd.DataFrame.from_dict(Customer_dict)
Customer_df
```

```python
#code goes here
#Deleting "age" column from Customer_df Dataframe
Customer_df1 = Customer_df.drop('age', axis=1)
Customer_df1
```

### Show the total amount of passengers who survived and not survived (10pts)

Untuk mengetahui jumlah penumpang selamat dan tidak selamat diperlukan perhitungan penumpang berdasarkan pengkategorian data kolom 'Survived'. Pada dataframe 'df' variabel kolom 'PassengerId' memiliki type data kategori. oleh karena itu untuk menjumlahkan banyaknya passengers digunakan operasi 'count'.

**Kesimpulan**

Dengan anggapan bahwa 'kategori 0' didefinisikan sebagai 'not survived' dan 'kategori 1' sebagai 'survived' maka dapat disimpulkan bahwa hasil menunjukkan terdapat 549 penumpang tidak selamat dan 342 selamat.

```python
#code goes here
df.groupby(['Survived']).agg({'PassengerId':['count']})
```
