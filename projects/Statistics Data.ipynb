**MEAN (RATA-RATA)**

Mean/rata-rata merupakan salah satu pengukuran statistik yang menunjukkan kecenderungan letak pusat data (ukuran pemusatan data). Mean/rata-rata dihitung dengan membagi jumlah semua nilai data dalam satu variabel dengan ukuran data/banyaknya nilai data variabel tersebut. Nilai mean dipengaruhi oleh semua nilai data sehingga ukuran pemusatan data menggunakan mean akan memiliki validitas yang kurang baik apabila ternyata dalam variabel terdapat nilai-nilai ekstrim.


```python
mean = np.mean(df)
print(f"Mean All Variabel =\n{mean}")
```

**VARIAN**

Varian merupakan salah satu ukuran statistik yang menunjukkan bagaimana sebaran dari data tersebut,bagaimana kondisi sebuah data menyebar di kelompok data. Varian menunjukkkan sebaran nilai data terhadap nilai mean. Semakin kecil nilai varians, semakin dekat sebaran data dengan rata-rata. Semakin besar nilai varian, semakin besar sebaran data terhadap nilai rata-ratanya. Varian biasanya menghasilkan nilai yang cenderung besar sehingga terkadang kurang merepresentasikan sebaran data. Umumnya, nilai varians ini dilanjutkan ke perhitungan standar deviasi yang merupakan akar dari nilai varians sehingga menghasilkan nilai yang cenderung kecil dan dapat dengan mudah menunjukkan representasi sebara data terhadap nilai mean.

```python
varian = np.var(df)
print(f"Varian All Variabel =\n{varian}")
```

```python
std = np.std(df)
print(f"Standar Deviasi All Variabel =\n{std}")
```

Berdasarkan output di atas variabel yang memiliki sebaran data tergolong tinggi adalah `'AGE'`, `'BP'`,`'S1'`, `'S2'`, `'S3'`,`'S6'`, dan `'Y'`. Sedangkan variabel lainnya memiliki nilai varian yang tergolong rendah.

### **Task 3: What is the the distribution of Age, Sex, BMI and Y variables?**

Untuk mengetahui distribusi dari suatu Variabel Data bertipe Numerik dapat menggunakan tampilan histogram sehingga dapat diketahui bentuk histogramnya untuk mengetahui distribusinya. Dapat juga dilihat dari nilai mean dan median. Apabila kedua ukuran statistik tersebut memiliki kecenderungan nilai yang sama maka dapat data mendekati distribusi normal.

Untuk menampilkan histogram dari data dapat digunakan:

code: ***sns.displot(data=namadataframe['variabel'])***

Sementara, pada data kategorik memiliki jenis distribusi yang berbeda. Macam distribusi data kategorik adalah:
1. Distribusi Binomial 

  Merupakan distribusi yang paling sederhana. Umumnya distribusi binomial akan digunakan untuk percobaan yang merupakan data kategorik yang memiliki 2 kemungkinan nilai data.
2. Distribusi Multinomial

  Distribusi multinomial sebenarnya merupakan pengembangan dari distribusi binomial. Jika pada distribusi binomial hanya terdapat dua kemungkinan yang akan terjadi, pada distribusi multinomial memungkinkan untuk terdapat lebih dari dua kemungkinan.
3. Distribusi Poisson

  Distribusi yang satu ini akan lebih banyak berurusan dengan peluang yang mungkin terjadi dalam rentang waktu tertentu atau dalam daerah yang telah dipilih. 


####**1. Variabel AGE**

**Distribusi dilihat dari Histogram**

```python
sns.displot(data=df['AGE'], kde=True)
```

Berdasarkan output di atas terlihat bahwa pasien diabates paling banyak berusia 50-55 tahun. Histogram tidak memiliki bentuk yang simetris dan cenderung left-skewness. Hal tersebut berarti data cenderung tidak berdistribusi normal.
